```python
# --- Import libraries ---
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, TensorBoard
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.applications.vgg16 import VGG16, preprocess_input
from keras.models import Model
from keras.layers import Input, Lambda, Dense, Flatten, Dropout, BatchNormalization
import tensorflow as tf
import cv2
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')


```

```python
# --- Setup dataset path ---
data_path = "C:/Users/Nana Casmana/Downloads/Garbage Classification-20241121T105258Z-001/Garbage Classification/dataset/inorganic-waste"  # Update this path
data_classes = os.listdir(data_path)
```

```python
# --- List classes and count images in each class ---
for class_ in data_classes:
    print(f"{class_} has {len(os.listdir(os.path.join(data_path, class_)))} images.")
```

```python
# --- Create DataFrame for image paths and labels ---
data = pd.DataFrame()
for class_ in data_classes:
    temp = pd.DataFrame()
    temp['path'] = [os.path.join(data_path, class_, img)
                    for img in os.listdir(os.path.join(data_path, class_))]
    temp['type_trash'] = class_
    data = pd.concat([data, temp], ignore_index=True)

data = data.sample(frac=1).reset_index(drop=True)  # Shuffle data
print(data.head(10))
```

```python
# --- Data visualization ---
plt.figure(figsize=(10, 6))
sns.countplot(data['type_trash'], palette="Set2")
plt.title('Image Count per Class')
plt.xticks(rotation=45)
plt.show()
```

```python
# --- Display sample image ---
def display_images_with_labels(df, idx):
    row = df.iloc[idx]
    img = mpimg.imread(row['path'])
    plt.imshow(img)
    plt.title(f"Label: {row['type_trash']}")
    plt.axis('off')
    plt.show()


display_images_with_labels(data, 0)
```

```python
# --- Split dataset ---
train_data_df, temp_data_df = train_test_split(
    data, test_size=0.2, stratify=data['type_trash'], random_state=42)
val_data_df, test_data_df = train_test_split(
    temp_data_df, test_size=0.5, stratify=temp_data_df['type_trash'], random_state=42)

print(f"Train: {len(train_data_df)}, Validation: {len(val_data_df)}, Test: {len(test_data_df)}")
```

```python
# --- ImageDataGenerator ---
BATCH_SIZE = 64

train_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
train_data = train_gen.flow_from_dataframe(
    dataframe=train_data_df,
    directory=None,
    x_col='path',
    y_col='type_trash',
    target_size=(224, 224),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

val_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_data = val_gen.flow_from_dataframe(
    dataframe=val_data_df,
    directory=None,
    x_col='path',
    y_col='type_trash',
    target_size=(224, 224),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
test_data = test_gen.flow_from_dataframe(
    dataframe=test_data_df,
    directory=None,
    x_col='path',
    y_col='type_trash',
    target_size=(224, 224),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)
```

```python
# --- Compute class weights ---
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(data['type_trash']),
    y=data['type_trash']
)
class_weights = {i: weight for i, weight in enumerate(class_weights)}
print("Class Weights:", class_weights)
```

```python
# --- Build the model ---
conv_base = VGG16(include_top=False, weights='imagenet',
                  input_shape=(224, 224, 3))
for layer in conv_base.layers[:-2]:
    layer.trainable = False

top_model = conv_base.output
top_model = Flatten()(top_model)
top_model = Dense(1024, activation='relu')(top_model)
top_model = BatchNormalization()(top_model)
top_model = Dropout(0.1)(top_model)
top_model = Dense(512, activation='relu')(top_model)
top_model = BatchNormalization()(top_model)
top_model = Dropout(0.1)(top_model)
output_layer = Dense(len(data_classes), activation='softmax')(top_model)

model = Model(inputs=conv_base.input, outputs=output_layer)
model.summary()

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# --- Callbacks setup ---
checkpoint = ModelCheckpoint(filepath='model_vgg16_best.keras',
                             save_best_only=True, monitor='val_accuracy', mode='max')
early_stop = EarlyStopping(
    monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
```

```python
# --- Train the model ---
result = model.fit(
    train_data,
    validation_data=val_data,
    epochs=20,
    class_weight=class_weights,
    callbacks=[checkpoint, early_stop, reduce_lr]
)
```

```python
# --- Plot results ---
plt.plot(result.history['accuracy'], label='Train Accuracy')
plt.plot(result.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title("Model Accuracy")
plt.show()

plt.plot(result.history['loss'], label='Train Loss')
plt.plot(result.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Model Loss")
plt.show()
```

```python
# --- Evaluate the model ---
predictions = model.predict(val_data)
predicted_labels = np.argmax(predictions, axis=1)
true_labels = val_data.classes

conf_matrix = confusion_matrix(true_labels, predicted_labels)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d',
            xticklabels=val_data.class_indices.keys(), yticklabels=val_data.class_indices.keys())
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

print(classification_report(true_labels, predicted_labels,
      target_names=list(val_data.class_indices.keys())))
```

```python
# --- Predict individual image ---
from tensorflow.keras.preprocessing.image import load_img, img_to_array


def predict_single_image(image_path, model, class_indices):
    classes = {v: k for k, v in class_indices.items()}
    img = load_img(image_path, target_size=(224, 224))
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    predictions = model.predict(img_array)
    predicted_label = np.argmax(predictions, axis=1)[0]
    return classes[predicted_label]


# Replace with your image path
image_path = "C:/Users/Nana Casmana/Downloads/Garbage Classification-20241121T105258Z-001/Garbage Classification/dataset/inorganic-waste/paper/R_3055.jpg"
predicted_class = predict_single_image(
    image_path, model, test_data.class_indices)
```
